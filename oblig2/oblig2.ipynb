```python
# DAFE1000 Oblig2

# Importerer bibliotek
import numpy as np
import matplotlib.pyplot as plt

# Definer funksjonen i oppgaven
def f(x):
    return np.exp(-x/4) * np.arctan(x)

# Definer funksjonen vi skal finne nullpunktet til (f'(x) uten eksponentialleddet). Nullpunktet tilsvarer der f'(x)=0
def g(x):
    return np.arctan(x) - 4/(x**2 + 1)

# Definer den deriverte av g(x) for bruk i Newtons metode. Utregning i notater.md
def g_derivert(x):
    return 1/(x**2 + 1) + (8*x)/(x**2 + 1)**2

# Initialisering av variabler for den numeriske løseren i løkken
x_n = 2.0          # Startgjett for x
toleranse = 1e-5   # Toleransegrense (hvor nøyaktig svar vi krever)
feil = 1           # Initialiserer feil-variabelen for å starte løkken
teller = 0         # Teller antall iterasjoner for å unngå evig løkke

# Newtons metode: repeterer beregningen til endringen er mindre enn toleransen h
while (feil > toleranse) and (teller < 100):
    # Oppdaterer x ved hjelp av Newtons formel: x - g(x)/g'(x)
    x_neste = x_n - g(x_n)/g_derivert(x_n)

    # Beregner den absolutte forskjellen mellom forrige og nåværende iterasjon
    feil = np.abs(x_neste - x_n)

    # Oppdaterer verdien for neste runde i løkken
    x_n = x_neste
    teller += 1
```
